```python
def tracking_ann(returns):
    summary=returns.agg(["mean", "std"]).T
    summary["Premium"]=summary["mean"]*252
    summary["TE"]=summary["std"]*np.sqrt(252)
    summary.drop(columns=["mean", "std"], inplace=True)
    return summary
    
```
